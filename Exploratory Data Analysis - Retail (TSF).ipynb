```python
profit= round(df['Profit'].sum(),2)
print('Total Profit is',profit)
```

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values()
print(category_profit)

plt.bar(category_profit.index,category_profit.values)
plt.title("Profit & loss on Category")
plt.ylabel("Profit/Loss")
plt.show()
```

#### For above bar plot we can see that Technology having more profit and Furniture having low profit.

```python
sub_category_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()
print(sub_category_profit)

plt.figure(figsize=(13,7))
plt.bar(sub_category_profit.index,sub_category_profit.values)
plt.title("Profit & loss on Sub-Category")
plt.ylabel("Profit/Loss")
plt.xticks(rotation=90)
plt.show()
```

#### For above bar plot we can see that Tables are in loss and Copiers are in profit

```python
Region_profit = df.groupby('Region')['Profit'].sum().sort_values()
print(Region_profit)

plt.figure(figsize=(7,5))
plt.bar(Region_profit.index,Region_profit.values)
plt.title("Profit & loss on Region")
plt.ylabel("Profit/Loss")
plt.show()
```

#### For above bar plot we can see that West region have more Profit and Central region have less Profit

```python
sns.barplot(x='Category',y='Sales',data=df)
```

#### For above bar plot we can see that Technology have more sales and office supplies have low sales

```python
region_sales = df.groupby("Region")['Sales'].sum().sort_values()
plt.pie(region_sales.values,labels=region_sales.index,autopct='%1.1f%%')
plt.title("Sales on Region")
plt.show()
```

#### For above Pie plot we can see that West region contains 31.6% sales thats why their is more profit compare to other

```python
Category_sales = df.groupby("Category")['Sales'].sum().sort_values()
plt.pie(Category_sales.values,labels=Category_sales.index,autopct='%1.1f%%')
plt.title("Sales on Category")
plt.show()
```

```python
discout_profit = df.groupby("Discount")['Profit'].sum().sort_values()
plt.figure(figsize=(8,5))
discout_profit.plot(kind='bar')
plt.title("Profit & loss on Discount")
plt.xlabel("Discount")
plt.ylabel("Profit/Loss")
plt.show()
```

#### From the above Bar plot we can say that discount is inversly proportional to profit i.e if discount is increased profit is decreased and loss increased and viseversa.

```python
shipmode_profit_sales = df.groupby("Ship Mode")[['Sales','Profit']].sum()
shipmode_profit_sales.plot(kind='bar')
plt.title("Profit & Sales on Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel('Amount')
plt.show()
```

####  From the above Bar plot we can say that Profit is directly proportional to Sales as Sales increase Profit also increase and viseversa

```python
State_profit_sales = df.groupby("State")[['Sales','Profit']].sum()
plt.figure(figsize=(15,5))
State_profit_sales.plot(kind='bar')
plt.title("Profit & Sales on State")
plt.xlabel("State")
plt.ylabel('Amount')
plt.show()
```

#### From the above Bar plot we can say that California and New York state has maximum sales and maximum profit as compared to other and there are some state having loss also i.e Texas.
